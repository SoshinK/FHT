```python
def test(point1, point2):
    channels = 3
    length = height = 64
    canvas = np.array([ [ [0] * channels for i in range(length)] for j in range(height)])
    canvas[point1[0]][point1[1]] = (255, 255, 255)
    canvas[point2[0]][point2[1]] = (255, 255, 255)
    line = fht.get_diadln(length, point1, point2)
    print("Points:", point1, point2)
    print("Found line: ", line)
    answer = fht.draw_vert_r(canvas, line[0], line[1], (193, 0, 32))
    plt.imshow(answer)
```
